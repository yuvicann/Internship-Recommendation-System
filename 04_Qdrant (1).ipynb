```python
# Define your query
query_text = "Action movie starring Tom Cruise released in 2022"

# Generate query embedding
query_vector = model.encode(query_text).tolist()
```

```python
results = qdrant_client.search(
    collection_name="movies",
    query_vector=query_vector,
    limit=5  # Number of recommendations to retrieve
)

# Display the results
for result in results:
    print("Title:", result.payload["title"])
    print("Director:", result.payload["director"])
    print("Stars:", result.payload["stars"])
    print("Category:", result.payload["category"])
    print("Release Year:", result.payload["release_year"])

    print()
```

```python
# Define genres, actors, and decades
genres = ["Action", "Drama", "Comedy", "Thriller", "Adventure"]
actors = ["Tom Cruise", "Leonardo DiCaprio", "Scarlett Johansson", "Christian Bale", "Jennifer Lawrence"]
decades = ["2000s", "2010s"]

# Generate queries with combinations of genres, actors, and decades
queries = []
for i in range(len(genres)):
    for j in range(i + 1, len(genres)):
        for actor in actors:
            for decade in decades:
                queries.append(f"{genres[i]} and {genres[j]} movie starring {actor} from the {decade}")
```

```python
# Display total number of queries
total_queries = len(queries)
print(f"Total Queries: {total_queries}")

# Display first 10 queries
print("\nFirst 10 Queries:")
for i, query in enumerate(queries[:10], 1):
    print(f"{i}: {query}")
```

```python
import time
import pandas as pd

# Placeholder for results
results = []

# Loop over generated queries
for query in queries:
    start_time = time.time()

    # Encode the query
    query_vector = model.encode(query).tolist()

    # Perform search in Qdrant
    response = qdrant_client.search(
        collection_name="movies",
        query_vector=query_vector,
        limit=5  # Number of recommendations to retrieve
    )

    # Calculate latency
    latency = time.time() - start_time

    # Process recommendations
    recommendations = [
        hit.payload['title'] for hit in response
    ]

    # Save query results
    results.append({
        "Query": query,
        "Recommendations": ", ".join(recommendations),
        "Latency (s)": latency
    })
```

```python
# Create DataFrame
results_df = pd.DataFrame(results)

# Save to Excel
output_file = "decade_genre_actor_recommendations.xlsx"
results_df.to_excel(output_file, index=False)

print(f"Results saved to {output_file}")
```

```python
import random
from itertools import combinations

# Define seed query parameters
genres = data['Category'].dropna().unique().tolist()  # Unique genres
popular_actors = [
    "Tom Cruise", "Jennifer Lawrence", "Leonardo DiCaprio", "Scarlett Johansson",
    "Brad Pitt", "Natalie Portman", "Christian Bale", "Anne Hathaway",
    "Robert Downey Jr.", "Emma Stone"
]  # Top popular actors
decades = ["2000s", "2010s"]  # Specific decades
language = "English"  # Restrict to English language

# Generate Level 4 seed queries
seed_queries = []
for genre_combo in combinations(genres, 2):  # Two-genre combinations
    for actor in popular_actors:
        for decade in decades:
            seed_queries.append(f"{genre_combo[0]}  movie starring {actor} from the {decade} in {language}.")

# Shuffle and select 200 diverse queries
random.shuffle(seed_queries)
seed_queries = seed_queries[:200]

# Display some sample queries
print("Sample Seed Queries:\n")
for i, query in enumerate(seed_queries[:10], start=1):  # Show the first 10 queries
    print(f"{i}: {query}")
print(f"\nTotal Seed Queries Generated: {len(seed_queries)}")
```

```python
from sentence_transformers import SentenceTransformer

# Initialize the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for the seed queries
query_embeddings = [model.encode(query).tolist() for query in seed_queries]
```

```python
import time

# Store recommendations and latency
recommendations = []
latency = []

# Perform vector search
for i, (query, embedding) in enumerate(zip(seed_queries, query_embeddings)):
    start_time = time.time()
    results = qdrant_client.search(
        collection_name="movies",
        query_vector=embedding,
        limit=5
    )
    end_time = time.time()


    recommendations.append({
        "query": query,
        "recommendations": [result.payload['title'] for result in results],
        "latency": round(end_time - start_time, 4)
    })


print("\nSample Recommendations:")
for rec in recommendations[:3]:
    print(f"Query: {rec['query']}")
    print(f"Recommendations: {rec['recommendations']}")
    print(f"Latency: {rec['latency']} seconds\n")
```

```python
import pandas as pd

# Prepare data for Excel
df_results = pd.DataFrame(recommendations)

# Save to Excel
output_file = "seed_query_recommendations.xlsx"
df_results.to_excel(output_file, index=False)
print(f"Results saved to {output_file}")
```
